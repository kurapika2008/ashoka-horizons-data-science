```python
#Q1
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('food-master-file.csv')
#top 5 entries
df.head(5)
```

```python
#bottom 5 entries
df.tail(5)
```

```python
#Q2 removing missing values
df = df.fillna("0")
df.head(3)
```

```python
skip_arr= ['Name (Primary Key)', "Meal Type", "Cuisine", "Best Cooking Method"]
for col in df.columns:
    if col in skip_arr:
        continue
    df[col] = df[col].astype(float)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
```

```python
#line chart
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(df['Calories (Approx)'].head(50), df['Serving Size (in gms)'].head(50), marker='o', color='#d1bdf2', markeredgewidth='2.0', markeredgecolor='purple', markerfacecolor='lavender', linestyle= ':')
plt.xlabel('Total Calories')
plt.ylabel('Serving Size (g)')
plt.title('Calories vs Serving Size Line Plot', fontsize=15, color='#9282e0')

plt.subplot(1,2,2)
plt.plot(df['Prep Time (in minutes)'].head(50), df['Cooking time (in minutes)'].head(50), marker='o', color='#f2cbf0', markeredgewidth='2.0', markeredgecolor='pink', markerfacecolor='#fce6fb', linestyle= '--')
plt.xlabel('Preparation Time (min)')
plt.ylabel('Cook Time (min)')
plt.title('Prep Time vs Cook Time Line Plot', fontsize=15, color='#b53ead')
```

```python
#bar plot
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.bar(x=df['Cuisine'].unique(), height=df['Cuisine'].value_counts().to_list(), color=['#b6e1f2', '#b6bff2', '#6890a1', '#6869a1'])
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Count of Cuisine', fontsize=15, color='#30436b')
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light'
)

plt.subplot(1,2,2)
plt.bar(x=df['Meal Type'].unique(), height=df['Meal Type'].value_counts().to_list(), color=['#bef7c5', '#779e65', '#8fdbb6', '#478064'])
plt.xlabel('Meal Type)')
plt.ylabel('Count')
plt.title('Count of Meal Type', fontsize=15, color='#2f694d')
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light'
)
plt.show()
```

```python
#area chart
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(df['Calories (Approx)'].head(15), df['Serving Size (in gms)'].head(15), marker='o', color='orange', markeredgewidth='2.0', markeredgecolor='orange', markerfacecolor='#fce8d4', linestyle= ':')
#plt.fill_between(df['Calories (Approx)'].head(15), df['Serving Size (in gms)'].head(15), color='#ffdbba')
plt.xlabel('Total Calories')
plt.ylabel('Serving Size (g)')
plt.title('Calories vs Serving Size Area Plot', fontsize=15, color='#d9873b')
plt.xticks(
    rotation=90, 
    horizontalalignment='right',
    fontweight='light'
)

plt.subplot(1,2,2)
plt.plot(df['Prep Time (in minutes)'], df['Cooking time (in minutes)'], marker='o', color='red', markeredgewidth='2.0', markeredgecolor='red', markerfacecolor='#fac3c3', linestyle= '--')
plt.fill_between(df['Prep Time (in minutes)'], df['Cooking time (in minutes)'], color='#fac3c3')
plt.xlabel('Prep Time (min)')
plt.ylabel('Cook Time Time (min)')
plt.title('Prep Time vs Cook Time Area Plot', fontsize=15, color='#cc5050')
```

```python
#dendogram
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster
from scipy.spatial.distance import pdist
```

```python
df_plot= df.drop(columns=['Name (Primary Key)', 'Cuisine', 'Meal Type', 'Best Cooking Method'], axis=1)

plt.figure(figsize=(10,5))
link = linkage(df_plot, 'ward')
dendrogram(link, color_threshold=200, truncate_mode='lastp', p=10)
plt.xticks(rotation=90)
plt.title('Dendrogram Plot', fontsize=20, color='orange')
plt.show()
```
